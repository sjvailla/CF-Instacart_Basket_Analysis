```python
df = df_ord_prod_merged[:1000000]
```

```python
#Define price label
def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
#Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
#Step 2
# Change Busiest day to Busiest days. Now need the 2 busies days (0,1) and the least busy (3,4).
result_2 = []

for value in df_ord_prod_merged["orders_day_of_week"]:
    if value == 0 or value == 1:
        result_2.append("Busiest day")
    elif value == 4 or value == 3:
        result_2.append("Least busy")
    else:
        result_2.append("Regularly busy")
```

```python
df_ord_prod_merged['busiest_days'] = result_2
```

```python
#Step 3
#Check values
df_ord_prod_merged['busiest_days'].value_counts(dropna = False)
```

```python
#Step 4
#ID the busiest hours of the day. Rather by hour, they want periods of time labeled Most ordered, Average orders, and Fewest orders.
#Make a new column labeled busiest_period_of_day.
df_ord_prod_merged['order_hour_of_day'].value_counts(dropna = False)
```

Because this is based on 24 hours, I am breaking it into the 3 and will be puting the values into their respective 8 hours sections.

```python
#For loop of the hours of day
hours_of_day = []

for value in df_ord_prod_merged["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        hours_of_day.append("Most ordered")
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        hours_of_day.append("Fewest orders")
    else:
        hours_of_day.append("Average orders")
```

```python
#Adding for loop
df_ord_prod_merged['busiest_period_of_day'] = hours_of_day
```

```python
#Checking addition above
df_ord_prod_merged['busiest_period_of_day'].value_counts(dropna = False)
```
