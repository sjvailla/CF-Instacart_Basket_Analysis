```python
#Step 4
# Create a loyalty flag for existing customers using transform and loc
ord_prod_merged['max_order'] = ord_prod_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Accuracy check
pd.options.display.max_rows = None
```

```python
ord_prod_merged.head(30)
```

```python
#Create a loyalty flag with the following values
#If the max order is over 40, they are a Loyal customer
#If the max order is between 10 and 40, they are a Regular customer
#If the max order is under 10, they are a New customer
ord_prod_merged.loc[ord_prod_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ord_prod_merged.loc[(ord_prod_merged['max_order'] <= 40) & (ord_prod_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ord_prod_merged.loc[ord_prod_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Frequency check
ord_prod_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
#step 5
#Find the mean in spending in the loyalty flag grou
ord_prod_merged.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
#Step 6
#Create a spending flag with the following criteria 
#If the mean price of products purchased are lower than 10, Low spender
#If the mean price of products purchased are higher than or equal to 10, High spender
ord_prod_merged['mean_price'] = ord_prod_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Accuracy check
ord_prod_merged.head(50)
```

```python
#Creating spending flag
ord_prod_merged.loc[ord_prod_merged['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ord_prod_merged.loc[ord_prod_merged['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#Frequency check
ord_prod_merged['spending_flag'].value_counts(dropna = False)
```

```python
#Step 7
#Make a frequency flag for user's ordering behavior with median
ord_prod_merged['median_days'] = ord_prod_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#creating frequency flag
ord_prod_merged.loc[ord_prod_merged['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ord_prod_merged.loc[(ord_prod_merged['median_days'] > 10) & (ord_prod_merged['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ord_prod_merged.loc[ord_prod_merged['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ord_prod_merged['frequency_flag'].value_counts(dropna = False)
```
