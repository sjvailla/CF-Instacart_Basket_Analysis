# 05 Creating profiles for customers

```python
#Creating age groups
#Finding min and max
high_act_cust['age'].min()
```

```python
high_act_cust['age'].max()
```

```python
high_act_cust.loc[high_act_cust['age'] <=29, 'age_group'] = '18-29'
high_act_cust.loc[(high_act_cust['age'] > 29) & (high_act_cust['age'] < 40), 'age_group'] = '30-39'
high_act_cust.loc[(high_act_cust['age'] > 39) & (high_act_cust['age'] < 50), 'age_group'] = '40-49'
high_act_cust.loc[(high_act_cust['age'] > 49) & (high_act_cust['age'] < 60), 'age_group'] = '50-59'
high_act_cust.loc[(high_act_cust['age'] > 59) & (high_act_cust['age'] < 65), 'age_group'] = '60-65'
high_act_cust.loc[high_act_cust['age'] >=65, 'age_group'] = '65 and older'
```

```python
#Verifying age group
high_act_cust['age_group'].value_counts()
```

```python
#Creating income groups
#Finding min and max
high_act_cust['income'].min()
```

```python
high_act_cust['income'].max()
```

```python
high_act_cust.loc[high_act_cust['income'] <= 38133, 'income_group'] = 'Low income'
high_act_cust.loc[(high_act_cust['income'] > 38133) & (high_act_cust['income'] < 114400), 'income_group'] = 'Middle income'
high_act_cust.loc[(high_act_cust['income'] > 114400) & (high_act_cust['income'] < 300000), 'income_group'] = 'Upper income'
high_act_cust.loc[high_act_cust['income'] >= 300000, 'income_group'] = 'High wealth'
```

```python
#Verify income groups
high_act_cust['income_group'].value_counts()
```

Income groups were based on US census data for 2023

```python
#Importing dept
dept_path = r'C:\Users\sjvai\Documents\07-23 Instacart Basket Analysis'
```

```python
dept = pd.read_csv(os.path.join(dept_path, '02 Data', 'Prepared Data', 'departments_wrangled.csv'), index_col = False)
```

```python
dept
```

```python
dept.rename(columns = {'Unnamed: 0' : 'department_id'}, inplace = True)
```

```python
dept
```

```python
high_act_cust_merge = high_act_cust.merge(dept, on = 'department_id')
```

```python
high_act_cust_merge.head()
```

```python
#Creating dependant type

mask = (high_act_cust_merge['number_of_dependants'] > 0) & (high_act_cust_merge['department'] == 'babies')
high_act_cust_merge['dependant_type'] = high_act_cust_merge.groupby('user_id')['user_id'].transform(
    lambda x: 'has kids' if any(mask[x.index]) else 'has no kids'
)
```

```python
high_act_cust_merge.head(20)
```

```python
high_act_cust_merge['dependant_type'].value_counts()
```

```python
high_act_cust_merge['marital_status'].value_counts()
```

```python
#Creating family types

high_act_cust_merge.loc[(high_act_cust_merge['marital_status'].isin(['single', 'divorced/widowed', 'living with parents and siblings'])) & (high_act_cust_merge['number_of_dependants'] == 0), 'family_type'] = 'Single no dependants'
high_act_cust_merge.loc[(high_act_cust_merge['marital_status'].isin(['single', 'divorced/widowed', 'living with parents and siblings'])) & (high_act_cust_merge['number_of_dependants'] > 0), 'family_type'] = 'Single with dependants'
high_act_cust_merge.loc[(high_act_cust_merge['marital_status'] == 'married') & (high_act_cust_merge['number_of_dependants'] == 1), 'family_type'] = 'Married no dependants'
high_act_cust_merge.loc[(high_act_cust_merge['marital_status'] == 'married') & (high_act_cust_merge['number_of_dependants'] > 1), 'family_type'] = 'Married with dependants'
```

```python
high_act_cust_merge['family_type'].value_counts()
```

```python
high_act_cust_merge.info()
```

```python
#Creating time period

high_act_cust_merge.loc[(high_act_cust_merge['order_hour_of_day'] < 7) & (high_act_cust_merge['order_hour_of_day'] > 21), 'time_period'] = 'Overnight'
high_act_cust_merge.loc[(high_act_cust_merge['order_hour_of_day'] <= 12) & (high_act_cust_merge['order_hour_of_day'] > 7), 'time_period'] = 'Morning'
high_act_cust_merge.loc[(high_act_cust_merge['order_hour_of_day'] <= 18) & (high_act_cust_merge['order_hour_of_day'] > 12), 'time_period'] = 'Afternoon'
high_act_cust_merge.loc[(high_act_cust_merge['order_hour_of_day'] <= 21) & (high_act_cust_merge['order_hour_of_day'] > 18), 'time_period'] = 'Evening'
```

```python
high_act_cust_merge['time_period'].value_counts(dropna = False)
```

```python
#Assigning days

high_act_cust_merge.loc[high_act_cust_merge['orders_day_of_week'] == 0, 'day'] = 'Saturday'
high_act_cust_merge.loc[high_act_cust_merge['orders_day_of_week'] == 1, 'day'] = 'Sunday'
high_act_cust_merge.loc[high_act_cust_merge['orders_day_of_week'] == 2, 'day'] = 'Monday'
high_act_cust_merge.loc[high_act_cust_merge['orders_day_of_week'] == 3, 'day'] = 'Tuesday'
high_act_cust_merge.loc[high_act_cust_merge['orders_day_of_week'] == 4, 'day'] = 'Wednesday'
high_act_cust_merge.loc[high_act_cust_merge['orders_day_of_week'] == 5, 'day'] = 'Thursday'
high_act_cust_merge.loc[high_act_cust_merge['orders_day_of_week'] == 6, 'day'] = 'Friday'
```
