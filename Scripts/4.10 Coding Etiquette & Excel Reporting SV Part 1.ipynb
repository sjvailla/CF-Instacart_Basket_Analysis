# 03 Regional data

```python
#Checking states
ord_prod_merged['state'].value_counts()
```

```python
#For loop
#Assigning regions

location = []

for state in ord_prod_merged['state']:
    if state in ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']:
        location.append('Northeast')
    elif state in ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']:
        location.append('Midwest')
    elif state in ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']:
        location.append('South')
    elif state in ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']:
        location.append('West')
    else:
        location.append('error')
```

```python
#Adding region column
ord_prod_merged['region'] = location
```

```python
#Checking region column for errors
ord_prod_merged['region'].value_counts()
```

```python
ord_prod_merged.head()
```

```python
#Checking regions and spending
#creating crosstab
crosstab = pd.crosstab(ord_prod_merged['region'], ord_prod_merged['spending_flag'], dropna = False)
```

# 04 Creating dataframe without low-active customers

```python
#Creating low activity flag
ord_prod_merged.loc[ord_prod_merged['max_order'] < 5, 'customer_activity'] = 'Low activity'
```

```python
#Creating high activity flag
ord_prod_merged.loc[ord_prod_merged['max_order'] >= 5, 'customer_activity'] = 'High activity'
```

```python
ord_prod_merged['customer_activity'].value_counts()
```

```python
#Creating low activity dataframe
low_active_cust = ord_prod_merged[ord_prod_merged['customer_activity'] == 'Low activity']
```

```python
#Creating high activity dataframe
high_active_cust = ord_prod_merged[ord_prod_merged['customer_activity'] == 'High activity']
```
