# 06 Create visuals to show distribution of profiles

```python
#Age group disto
#Setting colors

age_colors = ['powderblue', 'lightskyblue', 'cornflowerblue', 'royalblue', 'mediumblue', 'navy']
```

```python
#Creating age group visuals

age_group_distro = high_act_cust['age_group'].value_counts().sort_index().plot.bar(color = age_colors)
plt.title("Distribution of Age Groups")
plt.xlabel("Age Group")
```

```python
vis_path = r'C:\Users\sjvai\Documents\07-23 Instacart Basket Analysis'
```

```python
age_group_distro.figure.savefig(os.path.join(vis_path, '04 Analysis', 'Visualizations', 'age_group_distro.png'))
```

```python
#Making visuals of orders by age groups

age_group_orders = high_act_cust.groupby('age_group')['order_id'].nunique()
```

```python
orders_by_age = age_group_orders.plot.bar(color = age_colors)
plt.title("Number of Orders by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Orders")
plt.savefig(os.path.join(vis_path, '04 Analysis', 'Visualizations', 'orders_by_age.png'), bbox_inches='tight')
```

```python
orders_by_age.figure.savefig(os.path.join(vis_path, '04 Analysis', 'Visualizations', 'orders_by_age.png')
```

```python
#Making visual for orders by age groups by the days

age_day = high_act_cust.groupby(['orders_day_of_week', 'age_group'])['order_id'].nunique().reset_index()
```

```python
age_day_2 = age_day.pivot(index='orders_day_of_week', columns = 'age_group', values = 'order_id')
```

```python
dow_age_bar = age_day_2.plot.bar(color = age_colors)
plt.title(" Number of Orders by Day and Age Group")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Orders")
plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))
plt.savefig(os.path.join(vis_path, '04 Analysis', 'Visualizations', 'dow_age_bar.png'), bbox_inches='tight')
```

```python
dow_age_bar.figure.savefig(os.path.join(vis_path, '04 Analysis', 'Visualizations', 'dow_age_bar.png'))
```
